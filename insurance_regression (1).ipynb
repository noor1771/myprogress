```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import data
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

```python
X = insurance_data.loc[:,'age'].values.reshape(-1,1)
y = insurance_data.loc[:,'charges'].values
```

```python
#Fit and predict
charges_model = LinearRegression()
charges_model.fit(X,y)
```

```python
#Plot the data and model
y_pred = charges_model.predict(X)
plt.scatter(X,y,color = 'r')
plt.plot(X,y_pred,color ='g')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Predict an unknown value
unk_x = [[75]]

x_pred = np.append(X,unk_x).reshape(-1,1)
y_pred = charges_model.predict(x_pred)
plt.scatter(X,y,color = 'r')
plt.plot(x_pred,y_pred,color ='g')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print(f'Charges for a {unk_x} year old should be {charges_model.predict(unk_x)}')
```
